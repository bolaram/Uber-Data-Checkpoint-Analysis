# Convert Date/Time and add some useful column

```python
data_frame["Date/Time"] = data_frame["Date/Time"].map(pd.to_datetime)
```

```python
data_frame.head()
```

```python
data_frame["Date/Time"][0]
```

```python
# creating Date_of_month column as dom
def get_dom(dt):
    return dt.day

data_frame["DoM"] = data_frame["Date/Time"].map(get_dom)
```

```python
data_frame.tail()
```

# See, the DoM column is created

```python
def get_weekday(dt):
    return dt.weekday()

data_frame["Weekday"] = data_frame["Date/Time"].map(get_weekday)

def get_hour(dt):
    return dt.hour

data_frame["Hour"] = data_frame["Date/Time"].map(get_hour)
```

```python
data_frame.tail()
```

# Analysis

# 1.Analyze the Date of Month

```python
hist(data_frame.DoM, bins = 31, rwidth = .8, range = (0.7, 31.5), alpha = .7)
xlabel("Date of month")
ylabel("Frequency")
title("Frequency by Date of month")
```

we can analysis data by hist and bar function. But, by bar function there is also other facilities too. Using bar we can extract other data from that data such as sorting.

```python
#for k, rows in data_frame.groupby("DoM"):
    #print(k, rows)
        
    #print(k, len(rows))
    
# Or with pandas we can get the same result also
def count_rows(rows):
    return len(rows)

date = data_frame.groupby("DoM").apply(count_rows)
```

```python
date
```

```python
plot(date)
```

# We can change this y-axis 10000 to 0 by using bar function

```python
date.plot.bar(color = "#1F77B4", alpha = .7)
xlabel("Date of Month")
ylabel("Frequency")
title("Frequency vs Date of month")
```

# Sorting data

```python
date_sorted = date.sort_values()
date_sorted
```

```python
date_sorted.plot.bar(color = "#1F77B4", alpha = .7)
xlabel("Date of Month")
ylabel("Frequency")
title("Frequency vs Date of month")
;
```

# 2. Analyze the hour

```python
hist(data_frame.Hour, bins = 24, rwidth = 0.8, range = (0.8, 24.8), alpha = .7)
xlabel("HOUR")
ylabel("Frequency")
title("Frequency vs HOUR")
;
```

# 3. Analyze the weekday

```python
hist(data_frame.Weekday, bins = 7, rwidth = .8, range = (-.5, 6.5), alpha = .7)
xticks(range(7), "Mon Tue Wed Thu Fri Sat Sun".split())
xlabel("Weekday")
ylabel("Frequency")
title("Frequency vs weekday")
;
```

# Cross analysis (Hour & Day of week)

```python
by_cross_h_w = data_frame.groupby(" Weekday Hour".split()).apply(count_rows).unstack()
```

```python
by_cross_h_w
```

```python
sns.heatmap(by_cross_h_w)
;
```

Here you can easily see that in Tue, Wed, Thu days, the number of uber rides are pretty high, more than 8000. But in the weekends, means in the Sat and Sun days, number of uber rides are below 2000 as expected. Because peoples are staying home in the weekend, generally.
